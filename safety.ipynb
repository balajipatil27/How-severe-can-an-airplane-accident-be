```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('train.csv')
df
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data.info()
```

```python
Label_encoder = LabelEncoder()
train_data['Severity_enc'] = Label_encoder.fit_transform(train_data['Severity'])
```

```python
X = train_data.drop(columns=['Severity', 'Severity_enc', 'Accident_ID'])
y = train_data['Severity_enc']
```

```python
X_test = test_data.drop(columns=['Accident_ID'])
```

```python
X_train , X_val , y_train, y_val = train_test_split(X,y,test_size= 0.2)
```

```python
model = RandomForestClassifier(n_estimators=100 , random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_val)
```

```python
test_pred = model.predict(X_test)
test_label = Label_encoder.inverse_transform(test_pred)

submission_df = pd.DataFrame({
    'Accident_ID' : test_data['Accident_ID'],
    'Severity' : test_label
})
```

```python
submission_df.info()
```

```python
submission_df.to_csv('prediction.csv', index = False)
```
